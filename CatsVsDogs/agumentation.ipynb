```python
from keras.preprocessing import image
from tensorflow.keras.preprocessing.image import ImageDataGenerator
```

```python
Img = image.load_img("dogs_vs_cats\\test\dogs\dog.30.jpg" , target_size=(200,200))
```

```python
Img
```

```python
datagen = ImageDataGenerator(
        rotation_range=30,
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True,
        width_shift_range=0.2,
        height_shift_range=0.2,
        fill_mode='reflect')
```

```python
type(Img)
```

```python
img = image.img_to_array(Img)
img.shape
```

```python
input_batch_img = img.reshape(1, 200, 200, 3)
```

```python
i = 0
for batch in datagen.flow(input_batch_img, batch_size=1,
                            save_to_dir='AgumentatedImg/', save_prefix='dog', save_format='jpg'):
    i += 1
    if i > 5:
        break  # otherwise the generator would loop indefinitely
```
